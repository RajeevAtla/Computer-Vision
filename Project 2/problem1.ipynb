```python
import gymnasium as gym
import math
import random
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from collections import namedtuple, deque
from itertools import count

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# This line creates an instance of the CartPole environment using the gymnasium library. 
# This environment simulates a classic reinforcement learning problem where the goal is to balance a pole on a cart by applying forces left or right.
env = gym.make("CartPole-v1")

# set up matplotlib
is_ipython = 'inline' in matplotlib.get_backend()
if is_ipython:
    from IPython import display

plt.ion()

# This line sets the computation device to either a GPU (CUDA or MPS) or CPU, 
# depending on what is available on the machine. 
# Using GPU acceleration significantly speeds up training for neural networks.

# if GPU is to be used
device = torch.device(
    "cuda" if torch.cuda.is_available() else
    "mps" if torch.backends.mps.is_available() else
    "cpu"
)
```

```python
Transition = namedtuple('Transition',
                        ('state', 'action', 'next_state', 'reward'))


class ReplayMemory(object):

    def __init__(self, capacity):
        self.memory = deque([], maxlen=capacity)

    def push(self, *args):
        """Save a transition"""
        self.memory.append(Transition(*args))

    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)

    def __len__(self):
        return len(self.memory)
```

```python
class DQN(nn.Module):

    def __init__(self, n_observations, n_actions):
        super(DQN, self).__init__()
        self.layer1 = nn.Linear(n_observations, 128)
        self.layer2 = nn.Linear(128, 128)
        self.layer3 = nn.Linear(128, n_actions)

    # Called with either one element to determine next action, or a batch
    # during optimization. Returns tensor([[left0exp,right0exp]...]).
    def forward(self, x):
        x = F.relu(self.layer1(x))
        x = F.relu(self.layer2(x))
        return self.layer3(x)
```

```python
# BATCH_SIZE is the number of transitions sampled from the replay buffer
# GAMMA is the discount factor as mentioned in the previous section
# EPS_START is the starting value of epsilon
# EPS_END is the final value of epsilon
# EPS_DECAY controls the rate of exponential decay of epsilon, higher means a slower decay
# TAU is the update rate of the target network
# LR is the learning rate of the ``AdamW`` optimizer
BATCH_SIZE = 128
GAMMA = 0.99
EPS_START = 0.9
EPS_END = 0.05
EPS_DECAY = 1000
TAU = 0.005
LR = 1e-4

# Get number of actions from gym action space
n_actions = env.action_space.n

# This line resets the environment to an initial state and retrieves the initial observation. 
# It’s done at the beginning of each episode to start a new attempt at solving the environment.

# Get the number of state observations
state, info = env.reset()
n_observations = len(state)

policy_net = DQN(n_observations, n_actions).to(device)
target_net = DQN(n_observations, n_actions).to(device)
target_net.load_state_dict(policy_net.state_dict())

optimizer = optim.AdamW(policy_net.parameters(), lr=LR, amsgrad=True)
memory = ReplayMemory(10000)


steps_done = 0


def select_action(state):
    global steps_done
    sample = random.random()
    eps_threshold = EPS_END + (EPS_START - EPS_END) * \
        math.exp(-1. * steps_done / EPS_DECAY)
    steps_done += 1
    if sample > eps_threshold:
        with torch.no_grad():
            # t.max(1) will return the largest column value of each row.
            # second column on max result is index of where max element was
            # found, so we pick action with the larger expected reward.
            return policy_net(state).max(1).indices.view(1, 1)
    else:
        return torch.tensor([[env.action_space.sample()]], device=device, dtype=torch.long)


episode_durations = []


def plot_durations(show_result=False):
    plt.figure(1)
    durations_t = torch.tensor(episode_durations, dtype=torch.float)
    if show_result:
        plt.title('Result')
    else:
        plt.clf()
        plt.title('Training...')
    plt.xlabel('Episode')
    plt.ylabel('Duration')
    plt.plot(durations_t.numpy())
    # Take 100 episode averages and plot them too
    if len(durations_t) >= 100:
        means = durations_t.unfold(0, 100, 1).mean(1).view(-1)
        means = torch.cat((torch.zeros(99), means))
        plt.plot(means.numpy())

    plt.pause(0.001)  # pause a bit so that plots are updated
    if is_ipython:
        if not show_result:
            display.display(plt.gcf())
            display.clear_output(wait=True)
        else:
            display.display(plt.gcf())
```

```python
def optimize_model():
    if len(memory) < BATCH_SIZE:
        return
    transitions = memory.sample(BATCH_SIZE)
    # Transpose the batch (see https://stackoverflow.com/a/19343/3343043 for
    # detailed explanation). This converts batch-array of Transitions
    # to Transition of batch-arrays.
    batch = Transition(*zip(*transitions))

    # Compute a mask of non-final states and concatenate the batch elements
    # (a final state would've been the one after which simulation ended)
    non_final_mask = torch.tensor(tuple(map(lambda s: s is not None,
                                          batch.next_state)), device=device, dtype=torch.bool)
    non_final_next_states = torch.cat([s for s in batch.next_state
                                                if s is not None])
    state_batch = torch.cat(batch.state)
    action_batch = torch.cat(batch.action)
    reward_batch = torch.cat(batch.reward)

    # Compute Q(s_t, a) - the model computes Q(s_t), then we select the
    # columns of actions taken. These are the actions which would've been taken
    # for each batch state according to policy_net
    state_action_values = policy_net(state_batch).gather(1, action_batch)

    # Compute V(s_{t+1}) for all next states.
    # Expected values of actions for non_final_next_states are computed based
    # on the "older" target_net; selecting their best reward with max(1).values
    # This is merged based on the mask, such that we'll have either the expected
    # state value or 0 in case the state was final.
    next_state_values = torch.zeros(BATCH_SIZE, device=device)
    with torch.no_grad():
        next_state_values[non_final_mask] = target_net(non_final_next_states).max(1).values
    # Compute the expected Q values
    expected_state_action_values = (next_state_values * GAMMA) + reward_batch

    # Compute Huber loss
    criterion = nn.SmoothL1Loss()

    # This line calculates the loss using the Huber loss function (SmoothL1Loss) 
    # between the predicted Q-values (state_action_values) 
    # and the target Q-values (expected_state_action_values). 
    # This loss drives the network to learn the correct Q-values to improve decision-making.
    loss = criterion(state_action_values, expected_state_action_values.unsqueeze(1))


    # This line resets the gradients of the optimizer. 
    # This is a necessary step before backpropagating the loss because 
    # PyTorch accumulates gradients by default, 
    # and not resetting would lead to incorrect gradient updates.
    # Optimize the model
    optimizer.zero_grad()

    # This line computes the gradients of the loss with respect to the network parameters. 
    # It backpropagates the loss through the neural network to update the model’s weights during training.
    loss.backward()
    # In-place gradient clipping
    torch.nn.utils.clip_grad_value_(policy_net.parameters(), 100)
    optimizer.step()
```

```python
if torch.cuda.is_available() or torch.backends.mps.is_available():
    num_episodes = 600
else:
    num_episodes = 50

for i_episode in range(num_episodes):
    # Initialize the environment and get its state
    state, info = env.reset()
    state = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
    for t in count():
        # This line selects an action using a policy (likely epsilon-greedy). 
        # It is used to decide the agent's next step based on the current state, 
        # balancing between exploration and exploitation.
        action = select_action(state)

        # This line takes the chosen action in the environment and receives the results: 
        # the next observation, 
        # the reward received, 
        # and whether the episode ended (terminated or truncated). 
        # This is a core part of the agent-environment interaction loop.
        observation, reward, terminated, truncated, _ = env.step(action.item())
        reward = torch.tensor([reward], device=device)
        done = terminated or truncated

        if terminated:
            next_state = None
        else:
            next_state = torch.tensor(observation, dtype=torch.float32, device=device).unsqueeze(0)

        # This line stores the transition (state, action, next_state, reward
        # into the replay memory. 
        # Storing transitions helps in breaking temporal correlations during training by allowing 
        # the model to be trained on randomly sampled experiences.
        
        # Store the transition in memory
        memory.push(state, action, next_state, reward)

        # Move to the next state
        state = next_state

        # Perform one step of the optimization (on the policy network)
        optimize_model()

        # Soft update of the target network's weights
        # θ′ ← τ θ + (1 −τ )θ′
        target_net_state_dict = target_net.state_dict()
        policy_net_state_dict = policy_net.state_dict()
        for key in policy_net_state_dict:
            # This line performs a soft update of the target network using the weights of the policy network. 
            # The parameter TAU determines the rate at which the target network adopts the new weights, 
            # which helps stabilize training by slowly updating the target network.
            target_net_state_dict[key] = policy_net_state_dict[key]*TAU + target_net_state_dict[key]*(1-TAU)
        target_net.load_state_dict(target_net_state_dict)

        if done:
            episode_durations.append(t + 1)
            plot_durations()
            break

print('Complete')
plot_durations(show_result=True)
plt.ioff()
plt.show()
```

```python

# Performance metrics plotting function
def plot_performance_metrics(agent_rewards, labels, title):
    plt.figure(figsize=(10, 5))
    for rewards, label in zip(agent_rewards, labels):
        plt.plot(np.cumsum(rewards), label=label)
    plt.xlabel('Episode')
    plt.ylabel('Cumulative Reward')
    plt.title(title)
    plt.legend()

# Function to run an agent (either trained or random) for a given number of episodes
def run_agent(agent, env, num_episodes, is_random=False):
    rewards = []
    for i_episode in range(num_episodes):
        state, _ = env.reset()
        state = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
        total_reward = 0
        done = False
        while not done:
            if is_random:
                action = env.action_space.sample()
            else:
                action = agent(state).max(1)[1].view(1, 1) if agent else env.action_space.sample()
            state, reward, terminated, truncated, _ = env.step(action.item())
            state = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
            total_reward += reward
            done = terminated or truncated
        rewards.append(total_reward)
    return rewards

# Initialize the environment and device
env = gym.make("CartPole-v1")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Set the number of episodes for different agents
random_agent_episodes = 100
short_training_episodes = 100
long_training_episodes = 600

# Random Agent Performance
env.reset()
random_rewards = run_agent(None, env, random_agent_episodes, is_random=True)

# Short Training Agent (assuming a pre-trained agent for a small number of episodes)
short_training_rewards = run_agent(policy_net, env, short_training_episodes)

# Long Training Agent (assuming the agent has been trained over a large number of episodes)
long_training_rewards = run_agent(policy_net, env, long_training_episodes)

# Plot performance metrics
plot_performance_metrics(
    [random_rewards, long_training_rewards, short_training_rewards],
    ["Random Agent", "Long Training", "Short Training"],
    "Performance Comparison of Different Agents"
)

# Close the environment
env.close()

plt.savefig("problem4.png")
plt.show()
```
